### 위원회 자르기

```python
# '-'를 기준으로 쪼개서 두 개의 새로운 칼럼을 만듦
parliament_df[['main_committee', 'sub_committee']] = parliament_df['committee_name'].str.split('-', n=1, expand=True).copy() # n=1; 최대 한 번만 -를 기준으로 문자열을 나누겠다는 의미

# 결과 확인
parliament_df.head()
```

```python
parliament_df = parliament_df[['id', 'date', 'conference_number', 'meeting_name', 'question_number','main_committee', 'sub_committee', 'committee_name', 'meeting_number', 'session_number', 'agenda', 'law', 'qna_type', 'context', 'context_learn', 'questioner_ID', 'answerer_ID', 'answer.comment', 'answer.keyword', 'normalized_date', 'date_transformed']]
```

```python
parliament_df = parliament_df.sort_values(by = ['main_committee', 'date','question_number'])
parliament_df = parliament_df.reset_index(drop = True)

parliament_df()
```

```python
parliament_df.info()
```

```python
parliament_df = parliament_df.drop(columns = ['date_transformed','normalized_date'])
parliament_df.info()
```

```python
parliament_df.head(10)
```

```python
len(parliament_df['conference_number'].unique())
```

```python

# 'agenda'의 첫 문자가 같은 행을 그룹으로 묶기 위한 함수
def get_agenda_group(agenda):
    return agenda.split(".")[0]

# 'agenda_group' 생성
parliament_df['agenda_group'] = parliament_df['agenda'].apply(get_agenda_group)

# 그룹화하여 context와 answer.comment를 리스트로 결합
parliament_wrap = parliament_df.groupby(['id', 'agenda_group']).agg({
    'context': lambda x: list(x),
    'answer.comment': lambda x: list(x),
    'agenda': 'first',
    'date': 'first',
    'meeting_name': 'first',
    'main_committee': 'first'
}).reset_index()

# context와 answer.comment를 결합하여 context_full 생성
parliament_wrap['context_full'] = parliament_wrap.apply(lambda row: row['context'] + row['answer.comment'], axis=1)

# agenda_index 부여: id 별로 1, 2, 3 순서로 인덱스 생성
parliament_wrap['agenda_index'] = parliament_wrap.groupby('id').cumcount() + 1
parliament_wrap['agenda_index'] = parliament_wrap['id'].astype(str) + '-' + parliament_wrap['agenda_index'].astype(str)

# 필요한 칼럼만 선택
parliament_wrap = parliament_wrap[['id', 'date', 'meeting_name', 'main_committee', 'agenda', 'agenda_index', 'context_full']]
parliament_wrap
```
