```python
# 특정 법안의 bill_id에 따라 30일 내외(혹은 사용자 지정 범위)의 날짜로 필터링한 후, 'committee'가 같은지 확인하는 함수
def get_indices_within_days_by_bill_id(bill_id, law_df, sub_wrapped, days=30):
    # 입력된 bill_id에 해당하는 law_df 행의 'date'와 'committee' 값 가져오기
    law_row = law_df.loc[law_df['bill_id'] == bill_id]
    law_date = law_row['date'].values[0]
    law_committee = law_row['committee'].values[0]
    
    # 'date'가 law_date 기준으로 지정된 범위 내에 있는 행들 필터링
    mask = (sub_wrapped['date'] >= law_date - pd.Timedelta(days=days)) & (sub_wrapped['date'] <= law_date + pd.Timedelta(days=days))
    sub_filtered = sub_wrapped[mask]
    
    # 필터링된 행 중에서 'committee'가 같은 행을 다시 필터링
    sub_committee_filtered = sub_filtered[sub_filtered['main_committee'] == law_committee]
    
    # 'committee'가 같은 데이터가 없으면 메시지를 출력하고 날짜 범위에 해당하는 행만 반환
    if sub_committee_filtered.empty:
        print(f"담당 위원회에서 전후 {days}일 안에 논의된 기록이 없습니다.")
        return sub_filtered.index.tolist(), sub_filtered
    else:
        return sub_committee_filtered.index.tolist(), sub_committee_filtered


# 사용 예시:
# bill_id = 'want_to_find'  # 확인하고 싶은 bill_id
# indices, matching_rows = get_indices_within_days_by_bill_id(bill_id, law_df, sub_wrapped)

# 일치하는 행들을 출력
# print(matching_rows)


# 특정 법안의 bill_id에 따라 30일 내외의 sub_wrapped 행을 확인하는 함수
# def get_indices_within_30_days_by_bill_id(bill_id, law_df, sub_wrapped):
#     # 입력된 bill_id에 해당하는 law_df 행의 'date' 값 가져오기
#     law_date = law_df.loc[law_df['bill_id'] == bill_id, 'date'].values[0]
    
#     # sub_wrapped의 'date'가 law_date 기준으로 30일 전후에 해당하는지 확인하는 조건 생성
#     mask = (sub_wrapped['date'] >= law_date - pd.Timedelta(days=30)) & (sub_wrapped['date'] <= law_date + pd.Timedelta(days=30))
    
#     # 조건을 만족하는 행의 인덱스를 리스트로 변환하여 저장
#     indices = sub_wrapped[mask].index.tolist()
    
#     return indices, sub_wrapped.loc[indices]  # 인덱스 리스트와 해당하는 행 반환

# 사용 예시:
# bill_id = 'want_to_find'  # 확인하고 싶은 bill_id
# indices, matching_rows = get_indices_within_30_days_by_bill_id(bill_id, law_df, sub_wrapped)

# 일치하는 행들을 출력
# print(matching_rows)
```

### 위원회 자르기

```python
# '-'를 기준으로 쪼개서 두 개의 새로운 칼럼을 만듦
parliament_df[['main_committee', 'sub_committee']] = parliament_df['committee_name'].str.split('-', n=1, expand=True).copy() # n=1; 최대 한 번만 -를 기준으로 문자열을 나누겠다는 의미

# 결과 확인
parliament_df.head()
```

```python
parliament_df = parliament_df[['id', 'date', 'conference_number', 'meeting_name', 'question_number','main_committee', 'sub_committee', 'committee_name', 'meeting_number', 'session_number', 'agenda', 'law', 'qna_type', 'context', 'context_learn', 'questioner_ID', 'answerer_ID', 'answer.comment', 'answer.keyword', 'normalized_date', 'date_transformed']]
```

```python
parliament_df = parliament_df.sort_values(by = ['main_committee', 'date','question_number'])
parliament_df = parliament_df.reset_index(drop = True)

parliament_df()
```

```python
parliament_df.info()
```

```python
parliament_df = parliament_df.drop(columns = ['date_transformed','normalized_date'])
parliament_df.info()
```

```python
# 'agenda'의 첫 문자가 같은 행을 그룹으로 묶기 위한 함수
def get_agenda_group(agenda):
    return agenda.split(".")[0]

# 'agenda_group' 생성
parliament_df['agenda_group'] = parliament_df['agenda'].apply(get_agenda_group)

# 그룹화하여 context와 answer.comment를 각각 리스트로 결합
parliament_wrap = parliament_df.groupby(['id', 'agenda_group']).agg({
    'context': lambda x: list(x),  # context를 리스트로 결합
    'answer.comment': lambda x: list(x),  # answer.comment를 리스트로 결합
    'agenda': 'first',
    'date': 'first',
    'meeting_name': 'first',
    'main_committee': 'first'
}).reset_index()

# context와 answer.comment를 리스트로 결합하여 context_full 생성
parliament_wrap['context_full'] = parliament_wrap.apply(lambda row: row['context'] + row['answer.comment'], axis=1)

# agenda_index 부여: id 별로 1, 2, 3 순서로 인덱스 생성
parliament_wrap['agenda_index'] = parliament_wrap.groupby('id').cumcount() + 1
parliament_wrap['agenda_index'] = parliament_wrap['id'].astype(str) + '-' + parliament_wrap['agenda_index'].astype(str)

# 필요한 칼럼만 선택
parliament_wrap = parliament_wrap[['id', 'date', 'meeting_name', 'main_committee', 'agenda', 'agenda_index', 'context_full']]
```
